```python
import xml.etree.ElementTree as ET
import pandas as pd

# Function to parse XML and convert to DataFrame
def parse_xml_to_df(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    all_records = []
    for child in root:
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
        all_records.append(record)

    return pd.DataFrame(all_records)

# Parse the XML file
xml_df = parse_xml_to_df('acm-sf-chapter-hackathon-small/small_product_data.xml')

# Now xml_df is a pandas DataFrame containing your XML data
```
